## **Design Proposal of a Large Language Model Application**

### **AI in the Arts:**

“Recent years have seen a rapid development in the use of digital technologies by arts, cultural and heritage organisations. This was accelerated by the Covid-19 pandemic, which drove artists to explore new ways of working and enabled wider audiences to access arts and culture online” (Gajjar & Jeary, 2024).

### **Initial Design Idea:**

Fans play an integral role in the success of artists within the music industry. Their engagement with media other than the songs themselves,  is just as essential in the promotion and relevance of the artists and their images.

Therefore, my proposal for an application of a large language model, is to construct a programme which possesses the ability to generate a creative writing piece, utilising song lyrics as inspiration.

This application would hopefully garner a successful reception as it portrays similarities to the phenomenon of ‘fanfiction’, however, it reduces the amount of unpaid fan labour required to create and distribute. “Fans have been seen as valuable prosumers who can support and promote their idols or other media product”, and this is why fan labour is so significant. (Derbaix & Korchia & Matis, 2023)

### **Interaction Design:**

In this proposal, the input data would consist of a sufficient library of song lyrics, which can be obtained from online sources such as Genius, after obtaining permission from the owner(s).

Here is the copyright license under which Genius operates:

ML Genius Holdings, LLC is fully licensed to display lyrics across all of its properties. In 2013, entered into licenses with every major music publisher: Sony/ATV Music Publishing, EMI Music Publishing, Universal Music Publishing Group, and Warner/Chappell Music. In addition, developed a form license with the National Music Publishers' Association (NMPA) which today covers more than 96% of the independent publisher market.

The machine learning model would then be required to learn and train itself on this data, in order to acquire a robust understanding of the way words are used to convey themes and storylines present within the music, and construct a creative writing piece that accurately adheres to such elements. To achieve this, the model should identify key characters, images and plots from the words provided.

I have concluded that the most efficient large language model for this particular application would be GPT-4. This is due to the fact that the model possesses the ability to accurately accept and emit text inputs. Furthermore, GPT-4 is declared to be more creative than models such as GPT-3.5, which would prove to be beneficial in this particular project of creative writing (openai.com, 2025).

When ChatGPT was queried on the relevant attributes of GPT-4, it stated that it is designed to:

-	Interpret style, tone, and emotion in text (like song lyrics).
-	Generate original stories, poems, or lyrics in a chosen voice or theme.
-	Maintain narrative structure and cohesive storytelling.
-	Detect sadness, joy, anger, nostalgia, etc. (helping the LLM steer the creative piece toward a similar or contrasting emotional vibe).

(ChatGPT, 2025)

This proposal obtained the idea to utilise a GPT variant from the example: AI-augmented Human Imagination: Student Projects in Generative AI and Cultural Collections (Maclean, 2024).

Humans would interact with the model by selecting a song that they would like to convert into a story, and present it to the AI which would generate and return to the user, a creative writing piece utilising key aspects of their chosen music.

After the model has obtained the relevant data, human users could then provide a prompt stating any elements they would like the creative piece to contain, to ensure any desirable tropes or character arcs are included in the output.

This prompt could include requesting that the creative piece be returned to the user in their chosen language – as large language models such as GPT-4 have been programmed with the ability to carry out this task. This can be achieved by providing the model with data consisting of literature in other languages, to train itself on. (If the model has been trained on multiple languages, then even the input lyrics could be presented in another language).

In result of this, humans can then share and discuss their output pieces within their selective fandom communities allowing them to develop connections with those who share their interests.

It is vastly important to ensure that the model solely takes inspiration from the original piece, to ensure that the final output adheres to the copyright laws declared by the original artist. This is discussed in more detail further on in the proposal.

### **Design Value:**
This application is designed predominantly for recreational purposes and therefore, does not possess a large quantity of learning value for human users. However, the model does contain the potential to educate humans through example, on how to analyse a piece of literature and extract the important themes and messages from within it – when endeavouring to use it as a frame of reference for their own writing. This may be beneficial to those who are in education, or who’s field of work is related to writing articles based on existing literature.

The main values present within this design however, centre around engagement and inclusion. The creation of supplementary content surrounding an artist’s work, is a highly attractive spectacle for their respective fandoms. Platforms such as Reddit and Discord, contain a plethora of forums which are designed to allow users to interact with each other in regards to specific topics and conversations. For example, there is a popular Reddit subsection titled ‘Katyheads’, which contains a multitude of Katy Perry fans who discuss relevant news, songs, and fan labour projects on a daily basis. This example is relevant to the trial of the model presented later on in the proposal.

Therefore, the ability to choose songs and receive a story based on the lyrics, encourages users to engage with their favourite artists on a deeper level, whilst simultaneously engaging with likeminded individuals.

Furthermore, the ability to select any song, and have the output returned to the user in their chosen language, creates a sense of inclusion within the project – as the model would cover a large range of genres, styles, and artists ensuring that there are options for any and all human users.

Users have the freedom to request that important issues are included in the creative writing piece, in order to spread awareness in an engaging form of media – further highlighting the inclusive values of the project. For example, members of oppressed communities could disseminate their stories in order to allow their voices to be heard - this feat is made possible by the scale on which social media is able to circulate information. This is also possible for members of the LGBTQ+ community, racially marginalised communities, neurodiverse communities and many more.

### **Critical Reflection:**

Overall this application would not pose a large quantity of negative social and cultural implications, as the model would allow anyone from any culture or social background to interact with it in a way that is meaningful to their own personal experiences. It can also allow people to spread awareness on prominent issues within society through the engaging format of song and storytelling. Conversely, it does have the potential to inaccurately portray or interpret the implied context of a song, having the opposite effect. This can be rectified by ensuring that the model is correctly trained on the themes embedded within the lyrics.

### **Ethical and Social/Cultural Issues:**

This design could potentially be utilised as a form of promotional media for artists within the industry. Using AI as a form of marketing and advertisement carries potential ethical risks. For example, the purpose of advertising is to create an authentic portrayal of a brand or item, and there is a concern that the use of artificial intelligence in these situations, may lead to “Eventual Homogenisation” (Iliff, 2024).

Fortunately, GPT-4 claims to possess the ability to add a more ‘human-level performance’ to its outputs, due to its advanced ability to deeply analyse the tone and style of the input data (openai.com, 2025).

There is an indisputable level of bias present within artificial intelligence, including discrimination against factors such as race, gender, social class and many more. Therefore, there is a risk that the output creative writing pieces inadvertently display this bias in their plots, which has the potential to cause harmful stereotypes to circulate through modern media:

	“The algorithmic bias is caused by biased behaviours in the real world. It is the most difficult type of bias because in a sense this algorithmic bias is ‘correct’ – the algorithm does what statistically it is supposed to do: it correctly mirrors and represents the real world” (Baer, 2019).

Therefore, this application should it be implemented, would be required to function on the hope that the input songs do not contain any of these harmful biases within their lyrics – so that the LLM does not learn and utilise them in its output data.

### **Copyright:**

As the majority of songs are copyrighted by the artists, it is vital to ensure that the model takes only inspiration from the song lyrics, as opposed to implementing quotes directly from the original work into the creative piece. This would need to be specified within the output. Therefore, it is important to “use generative AI for research, inspiration or fine-tuning, but not blindly use AI-generated texts, images or audio content” (Schwarz, 2024).

### **Trial Prompt:**

This example utilises ChatGPT to construct a creative story inspired by Katy Perry’s Firework and Roar, in order to test the proposal.

**Here is an example in English -**

1. Prompt: Please generate a short creative story inspired by the lyrics of Katy Perry’s ‘Roar’. Focus on themes of independence and personal growth in a small-town setting.

***Output: The Quiet One***

When June Harper turned seventeen, people still called her “the quiet one.”

She lived in a town so small it only had one gas station, a diner that closed on Sundays, and neighbours who knew what kind of cereal you ate before you did. In Willow Creek, people didn’t change - they just settled.

But June was starting to feel different. It happened slowly, like how thunder starts long before the storm.

It began with a notebook. Black cover, blank pages. A gift from her aunt who had once backpacked through Europe and came back with purple hair and stories that made June’s mum roll her eyes.

“You should write down your thoughts,” her aunt said. “They’re louder than you think.”

She did. At first, it was whispers - scribbled lines about how tired she was of being agreeable, of swallowing opinions like bad medicine. Then came full sentences. Chapters. Rants. Dreams. She filled pages with the version of herself no one in Willow Creek had met.

One day, Ms. Carter, her English teacher, found her notebook on a desk.

“I didn’t mean to snoop,” she said, handing it back with a soft smile. “But you have a voice in there, June. You should let people hear it.”

That week, June signed up for the school’s spring talent show. No one knew what she was going to do - not even her mum, who offered to help her bake cookies for the concession table.

On the night of the show, she stood under the harsh auditorium lights, heart galloping like a wild horse. Then she read her poem, about feeling invisible in a town full of eyes, about the anger of being underestimated, about learning to speak even when her voice trembled.

When she finished, the room was quiet - not because they didn’t get it, but because they finally did.

Then the applause came. Not thunderous. Not roaring. But steady. Real.

(ChatGPT, 2025)

**Here is an example in Spanish -**

2. Prompt: Please generate a short creative story inspired by the lyrics of Katy Perry’s ‘Firework’. Focus on themes of self-confidence and self-acceptance in a beach setting. Please display the result in Castilian Spanish.

***Output: La Hoguera***

La playa estaba llena aquella noche, pero Isla se sentaba sola, con los tobillos enterrados en la arena fresca y la sudadera cerrada hasta el cuello. A lo lejos, la luz de una hoguera parpadeaba donde sus compañeros reían, sus rostros iluminados como estrellas. Nadie se había dado cuenta de que ella se había apartado, y la verdad, ya estaba acostumbrada.

Siempre se había sentido… fuera de lugar. Como la marea que se retira mientras todos los demás rompen contra la orilla. Demasiado callada. Demasiado rara. Demasiado.

Miró hacia abajo, hacia la concha que había recogido antes, acariciando con el pulgar su forma en espiral. Rota por un lado. Aun así, hermosa. Aun así, completa.

Una voz rompió la oscuridad.

¿Te importa si me siento?

Era Maren—descalza, con el pelo revuelto por el viento y una chispa en los ojos. Isla se encogió de hombros y se hizo un poco a un lado.

Se sentaron en silencio, ese tipo de silencio que no necesita ser llenado. Ese que hace sentir que, quizá, no eres invisible. Solo estás esperando ser visto.

¿Por qué no estás allí? —preguntó Maren en voz baja, señalando la hoguera con la barbilla.

Isla dudó.

No sé. Pensé que nadie lo notaría.

Maren la miró como si le estuviera faltando algo obvio.

Pero si tú eres la razón por la que estamos aquí. El cartel, la música… tú organizaste toda la noche.

Isla parpadeó.

Ya, pero eso es… detrás de las cámaras.

Exacto, respondió Maren. Tú eres la chispa. Tú encendiste todo esto.

Isla contempló el fuego a lo lejos. Desde ahí, parecía un fuego artificial, parpadeando contra la oscuridad. Algo pequeño que cambia el cielo.

Y por primera vez, lo creyó. Que ella importaba. Que no tenía que cambiar para encajar. Solo tenía que aparecer.

Se puso en pie, sacudiéndose la arena de los vaqueros.
Vamos - dijo con una sonrisa. Vamos a encender unas bengalas.

(ChatGPT, 2025)


(English Version for Translation Purposes)

"The Bonfire"

The beach was crowded that night, but Isla sat alone, ankles buried in the cool sand, hoodie zipped tight around her. Firelight flickered in the distance where her classmates laughed around a growing bonfire, their faces glowing like stars. They hadn’t noticed her slip away, and honestly, she was used to that.

She had always felt… off. Like the tide that pulled away while everyone else crashed forward. Too quiet. Too weird. Too much.

She looked down at the seashell she’d picked up earlier, running her thumb over its spiral shape. Broken on one side. Still beautiful. Still whole.

A voice broke through the dark.

“Mind if I sit?”

It was Maren - barefoot, with windblown hair and a spark in her eyes. Isla shrugged, scooting over slightly.

They sat in silence, the kind that didn’t need to be filled. The kind that made Isla feel like maybe she wasn’t invisible, just… waiting to be seen.

“Why aren’t you over there?” Maren asked gently, nodding toward the bonfire.

Isla hesitated. “I don’t know. Just... didn’t think anyone would notice.”

Maren looked at her like she was missing something obvious. “But you’re the reason we’re even here. That flyer you designed? The playlist? You planned the whole night.”
Isla blinked.

“Yeah, but that’s just... behind-the-scenes stuff.”

“Exactly,” Maren said. “You’re the spark. You lit the whole thing.”

Isla stared at the fire in the distance. It did look kind of like a firework from here—flickering bright against the night. Something small that changed the sky.

And for the first time, she believed it. That she mattered. That she didn’t need to change to fit in - she only needed to show up.

She stood, brushing the sand from her jeans.

“Come on,” she said, smiling. “Let’s go light some sparklers.”

(ChatGPT, 2025)
